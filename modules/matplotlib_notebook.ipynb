```python
x = np.arange(100)

fig, axs = plt.subplots(2,2)
style.use("classic")

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sign Wave")

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("Cosign Wave")

axs[1,0].plot(x, np.log(x))
axs[1,0].set_title("Log Function")

axs[1,1].plot(x, np.random.random(100))
axs[1,1].set_title("Random Function")

fig.suptitle("Four plots")
plt.tight_layout()
plt.savefig("four_plot.png", dpi=300, transparent= False)
plt.show()


```

```python
ax = plt.axes(projection = "3d")
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x,y,z)
ax.set_xlabel("x-labels")
ax.set_title("3D plot")
```

```python
ax = plt.axes(projection = "3d")
x = np.arange(0,50, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.scatter(x,y,z)
ax.set_xlabel("x-labels")
ax.set_title("3D plot")
```

```python
ax = plt.axes(projection = "3d")
x = np.arange(0,50, 0.1)
y = np.arange(0,50, 0.1)
z = np.cos(x+y)

ax.scatter(x,y,z)
ax.set_xlabel("x-labels")
ax.set_title("3D plot")
```

```python
ax = plt.axes(projection="3d")

x = np.arange(-5,5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)


surf = ax.plot_surface(X, Y, Z, cmap="Spectral")

ax.set_title("3D plot")
ax.view_init(azim=0, elev=90)
plt.show()
```
